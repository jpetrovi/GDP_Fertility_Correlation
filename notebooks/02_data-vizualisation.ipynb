```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

PROCESSED_DATA_FILENAME = '../data/processed/gdp_fertility.csv'

FIGURE_FILENAME = '../reports/figures/correlation.png'

df = pd.read_csv(PROCESSED_DATA_FILENAME, low_memory=False)

fig, ax = plt.subplots()
fig.set_size_inches(24, 14, forward=True)
ax.scatter(x = df['gdp'], y = df['fertility'])

fig.savefig(FIGURE_FILENAME, dpi=fig.dpi, bbox_inches='tight')

```
